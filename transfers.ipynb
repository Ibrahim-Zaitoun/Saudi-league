```python
def standardize_transfer_cost(transfer_cost):
    if transfer_cost is None:
        return None  # or any other value you prefer for missing data
    elif isinstance(transfer_cost, str):
        if 'loan' in transfer_cost.lower():
            return 'Loan'
        elif 'free transfer' in transfer_cost.lower():
            return 'Free Transfer'
        else:
            # Extract the numeric part and convert to euros
            if '€' in transfer_cost:
                transfer_cost = transfer_cost.replace('€', '').replace('m', 'e6').replace('k', 'e3').replace(',', '')
            return transfer_cost
    else:
        return transfer_cost

spl_transfers['fee'] = spl_transfers['fee'].apply(standardize_transfer_cost)
```

```python
spl_transfers['fee'].unique()
```

```python
def extract_numeric_value(transfer_cost):
    if transfer_cost is None or transfer_cost in ['Loan', 'Free Transfer','?','-']:
        return None
    else:
        return float(transfer_cost)

spl_transfers['fee_numeric'] = spl_transfers['fee'].apply(extract_numeric_value)
```

```python
spl_transfers.iloc[spl_transfers['fee_numeric'].nlargest(20).index].sort_values('fee_numeric')
```

```python
avg_transfers = spl_transfers.groupby('club_name')['fee_numeric'].mean()
avg_transfers = avg_transfers.reset_index()
avg_transfers.rename(columns={'fee_numeric': 'average_fee_numeric'},inplace= True)
avg_transfers
```

```python
plt.figure(figsize=(14,9))
sns.barplot(data= avg_transfers,y= 'club_name', x='average_fee_numeric')
plt.show()
```

```python
count_transfers = spl_transfers.groupby('club_name')['fee_numeric'].count().sort_values(ascending=False)
count_transfers = count_transfers.reset_index()
count_transfers.rename(columns={'fee_numeric': 'count_transfers'},inplace= True)
count_transfers
```

```python
plt.figure(figsize=(14,9))
sns.barplot(data= count_transfers,y= 'club_name', x='count_transfers')
plt.show()
```

```python
sum_transfers = spl_transfers.groupby('club_name')['fee_numeric'].sum().sort_values(ascending=False)
sum_transfers = sum_transfers.reset_index()
sum_transfers.rename(columns={'fee_numeric': 'sum_transfers'},inplace= True)
sum_transfers
```

```python
plt.figure(figsize=(14,9))
sns.barplot(data= sum_transfers,y= 'club_name', x='sum_transfers')
plt.show()
```
